
# Pandas

pandas is a python library used for data science and ML. in machine learning it is used for data preprocessing

```
# =============PANDAS=============
import pandas as pd

# KEY DATA STUCTURE IN PANDAS:
#       1. SERIES
#       2. DATAFRAME

#=============SERIES=============
# SERIES IS A LABELED ARRAY
s = pd.Series([1, 2, 3, 4])
print(s)
```

```
# You may also add labels
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
```
