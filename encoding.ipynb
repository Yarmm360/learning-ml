# Encoding in ML (Machine Learning)
### There are two types/ways of Encoding:
##### Label Encoding
##### One-Hot Endodin

```python
# ============================================================================
# ===============================LABEL-ENCODING===============================
# ============================================================================

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# DataFrame:
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Hours_Studied": [2, 5, 1, 7, 4, 3, 6, 2, 8, 5],
    "Attendance": [75, 90, 60, 95, 80, 70, 92, 68, 98, 85],
    "Previous_Score" : [58, 72, 45, 88, 65, 55, 81, 50, 93, 74],
    "Sleep_Hours": [6, 7, 5, 8, 6, 5, 7, 6, 8, 7],
    "Passed": ["No", "Yes", "No", "Yes", "Yes", "No", "Yes", "No", "Yes", "Yes"]
}

df = pd.DataFrame(data)

df_label = df.copy()
# LABEL-ENCODING:
le = LabelEncoder()
df_label['Passed_encoded'] = le.fit_transform(df_label['Passed'])
print("\nLabel Encoding data:")
print(df_label[['Student_ID', 'Hours_Studied', 'Attendance', 'Previous_Score', 'Sleep_Hours', 'Passed','Passed_encoded' ]])

print(df)
```

```python
# ============================================================================
# =============================ONE-HOT-ENCODING===============================
# ============================================================================

# DataFrame:
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'HR'],
    'Work_Type': ['Remote', 'Hybrid', 'Onsite', 'Remote', 'Hybrid'],
    'Experience_Yrs': [2, 5, 8, 3, 1]
}

df = pd.DataFrame(data)
print(df)

# ONE-HOT ENCODING:
df_one_hot_encoded = pd.get_dummies(df, columns=['Department'], dtype='int')
print(df_one_hot_encoded)
```
