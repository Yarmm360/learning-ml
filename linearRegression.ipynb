# Supervised Machine Learning
## Linear Regression:
Used to predict numerical data

```python
from sklearn.linear_model import LinearRegression

input_data = [[1], [2], [3], [4]] #input
output_data = [20, 40, 60, 80] # output

# =================================================================================================
# --------------|   LINEAR-REGRESSION   |----------------------------------------------------------
# =================================================================================================
model = LinearRegression()
model.fit(input_data, output_data) # model will learn from this data

hours = float(input("Enter hours: "))

predict_marks = model.predict([[hours]])
if predict_marks > 35:
    print("You may Pass, because...")
else:
    print("You may Fail, because...")
print(f"Based on {hours} hours, you may score: {predict_marks}")
```
