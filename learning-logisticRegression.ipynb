# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
x = [[1], [2], [3], [4], [5]]
y = [0, 0, 1, 1, 1] # "0" is fail, "1" is pass

# Logistic Regression Code
model = LogisticRegression()
model.fit(x, y) # giving model input + output to learn

hours = float(input("Enter Hours you study: ")) #taking input from user
predicted = model.predict([[hours]])[0]

if predicted == 1:
    print(f"You probably get passed on the basis of {predicted}")
else:
    print("You should study hard")
```
