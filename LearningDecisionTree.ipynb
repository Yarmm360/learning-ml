# Decision Tree

```python
from sklearn import tree

x = [[0, 0], [1, 1], [1, 2], [2, 1]]
y = [0, 0, 1, 1]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
clf_pred = clf.predict([[3, 4]])
print(clf_pred)
```

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
x, y = iris.data, iris.target

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
tree.plot_tree(clf)
```
