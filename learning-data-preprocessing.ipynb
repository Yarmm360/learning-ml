# Data Preprocessing

```python
import pandas  as pd

# creating dataset
data = {
    'Name': ['Sonu', None, 'Nonu', 'tonu', 'konu'],
    'Age': [None, 17, 18, None, 20],
    'Salary': [1000, 10000, 20000, 26000, None]
}

df = pd.DataFrame(data)
print(f"Orignal Data: \n{df}")

# To Get SUM OF All Missiiig Values in Data Set In Data:
print(df.isnull().sum())

# Fil Missing data:
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print(df)
```

```python
import pandas as pd

data = {
    'Name': ['Sonu', None, 'Nonu', 'tonu', 'konu'],
    'Age': [None, 17, 18, None, 20],
    'Salary': [1000, 10000, 20000, 26000, None]
}

df = pd.DataFrame(data)

# Converting DataFrame into floating point:
df['Age'] = df['Age'].astype(float)

# ERROR-FREE METHOD: Direct Assignment
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print(df)
```
