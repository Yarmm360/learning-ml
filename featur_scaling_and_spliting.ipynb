# Feature Scaling

```python
from narwhals import dataframe
from pandas.core.interchange import column
from sklearn.preprocessing import StandardScaler, minmax_scale, MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd

#====================================================================================================
# ----------------------------------|   Data-Frame      |--------------------------------------------
#====================================================================================================
data = {
    "StudyHours": [1, 2, 3, 4, 5, 6],
    "MarksObtained": [40, 84, 50, 70, 60, 52]
}
df = pd.DataFrame(data)

#====================================================================================================
# ----------------------------------|   Standared-Scaling      |-------------------------------------
#====================================================================================================
standared_scale = StandardScaler()
standered_scaled = standared_scale.fit_transform(df)

print("Scaler Output:")
print(pd.DataFrame(standered_scaled, columns=df.columns))

# minmax scaler:
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\n MinMax Scaler output:")
print(pd.DataFrame(minmax_scaled, columns=df.columns))

# Test-Train-Split
x = df[["StudyHours"]] # Input
y = df[["MarksObtained"]] # Output
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Test Data:")
print(x_test)
```
